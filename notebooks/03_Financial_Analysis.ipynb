# objective:

Analyze company performance, profitability, growth and valuation
across industries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("financial_clean_dataset (2).csv")
```

```python
df.head()
```

# Top Industries by Profit

```python
industry_profit = (
    df.groupby('Industry')['Net profit']
    .mean()
    .sort_values(ascending=False)
)

industry_profit.head(15)
```

# Top Industries by Revenue

```python
industry_sales = (
    df.groupby('Industry')['Sales']
    .mean()
    .sort_values(ascending=False)
)

industry_sales.head(15)
```

# Top Growth Industries

```python
industry_growth = (
    df.groupby('Industry')['Sales Growth %']
    .mean()
    .sort_values(ascending=False)
)

industry_growth.head(15)
```

# Top Companies by Market Cap

```python
top_marketcap = (
    df[['Name','Industry','Market Capitalization']]
    .sort_values(
        by='Market Capitalization',
        ascending=False
    )
)

top_marketcap.head(20)
```

# executice conclusion

 The analysis reveals a market characterized by strong sector concentration, particularly within energy and financial industries. While energy companies dominate revenue generation, financial institutions demonstrate superior profitability efficiency. Most companies operate with positive margins, although a significant proportion remain loss-making. Market valuation appears closely linked to profitability, but company size alone does not guarantee operational efficiency. Overall, the dataset highlights substantial variation in performance across industries, creating opportunities for benchmarking and strategic financial analysis.
