# Financial Analytics - Data Cleaning & Feature Engineering

## Objective

This notebook focuses on preparing the financial dataset for deeper analysis and dashboard development.

The main goals are:
- Select relevant financial variables
- Handle missing values
- Create financial KPIs
- Prepare a clean dataset for Power BI

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("Annual_P_L_1_final.csv")
df.head()
```

## Column Selection

Only the most relevant financial variables are selected for the analysis.

```python
selected_columns = [
    'Name',
    'NSE Code',
    'Industry',
    'Sales',
    'Sales preceding 12months', # Corrected
    'Net profit',
    'Net profit preceding 12months', # Corrected
    'EBIT',
    'EBIT last year', # Corrected
    'EBIDT last year', # Corrected
    'EPS',
    'OPM',
    'NPM last year',
    'Return on capital employed',
    'Market Capitalization'
]

df_clean = df[selected_columns].copy()

df_clean.head()
```

```python
missing_clean = pd.DataFrame({
    'Missing Values': df_clean.isnull().sum(),
    'Percentage': round((df_clean.isnull().sum() / len(df_clean)) * 100, 2)
})

missing_clean.sort_values(by='Percentage', ascending=False)
```

## Missing Values Treatment

Rows with missing values in essential business metrics such as Sales, Net Profit and Market Capitalization are removed.

For margin indicators, missing values are preserved when they do not affect the core analysis.

```python
essential_columns = [
    'Sales',
    'Net profit',
    'Market Capitalization'
]

df_clean = df_clean.dropna(subset=essential_columns)

df_clean.shape
```

## Feature Engineering

New financial indicators are created to support profitability, growth and valuation analysis.

```python
df_clean['Sales Growth %'] = (
    (df_clean['Sales'] - df_clean['Sales preceding 12months'])
    / df_clean['Sales preceding 12months']
) * 100
```

```python
df_clean['Net Profit Growth %'] = (
    (df_clean['Net profit'] - df_clean['Net profit preceding 12months'])
    / df_clean['Net profit preceding 12months']
) * 100
```

```python
df_clean['Market Cap to Sales'] = (
    df_clean['Market Capitalization'] / df_clean['Sales']
)
```

```python
df_clean['Market Cap to Net Profit'] = (
    df_clean['Market Capitalization'] / df_clean['Net profit']
)
```

```python
df_clean = df_clean.replace([np.inf, -np.inf], np.nan)
```

```python
df_clean.describe()
```

```python
def profitability_category(npm):
    if pd.isna(npm):
        return 'Unknown'
    elif npm < 0:
        return 'Loss-making'
    elif npm < 5:
        return 'Low Profitability'
    elif npm < 15:
        return 'Moderate Profitability'
    else:
        return 'High Profitability'

df_clean['Profitability Category'] = df_clean['NPM last year'].apply(profitability_category)
```

```python
df_clean['Profitability Category'].value_counts()
```

```python


df_clean.head()
```

```python
# Remove unrealistic margins

df_clean = df_clean[
    (df_clean['OPM'] > -100) &
    (df_clean['OPM'] < 100)
]

df_clean = df_clean[
    (df_clean['NPM last year'] > -100) &
    (df_clean['NPM last year'] < 100)
]
```

```python
df_clean[['OPM','NPM last year']].describe()
```

```python
df_clean.columns
```

```python
df_clean = df_clean.replace([np.inf, -np.inf], np.nan)

df_clean = df_clean[
    df_clean['Sales Growth %'].between(-500, 500, inclusive='neither') |
    df_clean['Sales Growth %'].isna()
]

df_clean = df_clean[
    df_clean['Net Profit Growth %'].between(-500, 500, inclusive='neither') |
    df_clean['Net Profit Growth %'].isna()
]

df_clean.shape
```

```python
df_clean[['Sales Growth %','Net Profit Growth %']].describe()
```

## Export Clean Dataset

The cleaned dataset is exported and will be used for SQL analysis and Power BI dashboard development.

```python
df_clean.to_csv("financial_clean_dataset.csv", index=False)
```
