## Data Quality Assessment

Before performing financial analysis, it is important to evaluate data completeness and identify potential quality issues.

```python
# ====================================
# DATA QUALITY CHECK
# ====================================
```

```python
missing = df.isnull().sum().sort_values(ascending=False)

missing[missing > 0]
```

```python
missing_percent = (df.isnull().sum()/len(df))*100

missing_percent.sort_values(ascending=False).head(20)
```

```python
df.duplicated().sum()
```

```python
industry_count = df['Industry'].value_counts()

industry_count.head(15)
```

```python
import matplotlib.pyplot as plt

industry_count.head(10).plot(
    kind='bar',
    figsize=(12,6),
    title='Top 10 Industries'
)

plt.show()
```

```python
# ====================================
# TOP COMPANIES BY SALES
# ====================================
```

```python
top_sales = df[['Name','Industry','Sales']]

top_sales = top_sales.sort_values(
    by='Sales',
    ascending=False
)

top_sales.head(10)
```

```python
top10 = top_sales.head(10)

plt.figure(figsize=(12,6))

plt.bar(
    top10['Name'],
    top10['Sales']
)

plt.xticks(rotation=45)

plt.title('Top 10 Companies by Sales')

plt.tight_layout()

plt.show()
```

```python
top_profit = df[['Name','Industry','Net profit']]

top_profit = top_profit.sort_values(
    by='Net profit',
    ascending=False
)

top10_profit = top_profit.head(10)

plt.figure(figsize=(12,6))

plt.bar(
    top10_profit['Name'],
    top10_profit['Net profit']
)

plt.xticks(rotation=45)

plt.title('Top 10 Companies by Net Profit')

plt.tight_layout()

plt.show()
```

```python
high_margin = df[
    ['Name','Industry','NPM last year']
]

high_margin = high_margin[
    high_margin['NPM last year'] > 0
]

high_margin.sort_values(
    by='NPM last year',
    ascending=False
).head(10)
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df['Sales'],
    df['Net profit']
)

plt.xlabel("Sales")

plt.ylabel("Net Profit")

plt.title("Sales vs Net Profit")

plt.show()
```

```python
# Missing Values

missing = pd.DataFrame({
    'Missing Values': df.isnull().sum(),
    'Percentage': round((df.isnull().sum()/len(df))*100,2)
})

missing = missing[
    missing['Missing Values'] > 0
]

missing.sort_values(
    by='Percentage',
    ascending=False
)
```

### Findings

The dataset presents relatively good quality.

Most variables have less than 15% missing values.

The highest concentration of missing values occurs in historical financial metrics such as Sales preceding 12 months and Net Profit preceding 12 months.

Core variables remain highly complete and suitable for analysis.

## Correlation Analysis

This section investigates relationships between key financial metrics and identifies factors associated with company profitability and market value.

```python
financial_columns = [
    'Sales',
    'Net profit',
    'EBIT',
    'EPS',
    'OPM',
    'NPM last year',
    'Market Capitalization'
]

correlation = df[
    financial_columns
].corr()

correlation
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    correlation,
    annot=True
)

plt.title("Financial Metrics Correlation")

plt.show()
```
